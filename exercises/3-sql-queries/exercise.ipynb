```python
# Load chinook dataset and query it using SQL magic into pandas dataframes
import pandas as pd
import sqlite3
%load_ext sql

# Load data
conn = sqlite3.connect("chinook.sqlite")

# Tell %sql about the database
%sql sqlite:///chinook.sqlite

# List tables in database
query = "SELECT name FROM sqlite_master WHERE type='table';"

# Read data into a Pandas DataFrame
tables = %sql $query

# Print head
display(tables)

# Query to get the first 5 rows of the `albums` table
result = %sql SELECT * FROM albums LIMIT 5;

# Display query result, note that Pandas DataFrame is returned!
display(result)
```

```python
%%sql
SELECT Name as Track_name, Title as Album_name
FROM tracks
LEFT JOIN albums ON tracks.AlbumId = albums.AlbumId
LIMIT 5;
```

```python
%%sql
SELECT DISTINCT country, COUNT (customerID)  AS Total_Customers
FROM customers
GROUP BY country
ORDER BY total_customers DESC
LIMIT 5;
```

```python
%%sql

SELECT DISTINCT g.name AS Genre, COUNT(t.trackid) AS Number_of_tracks
FROM genres g
LEFT JOIN tracks t ON g.genreid = t.genreid
GROUP BY g.name
ORDER BY Number_of_tracks DESC
LIMIT 5;
```

```python
%%sql
SELECT customers.firstname, customers.lastname, AVG(total) AS average_purchase
FROM customers
LEFT JOIN invoices ON invoices.customerid = customers.customerid
GROUP BY customers.customerid
ORDER BY average_purchase DESC
```

```python
%%sql
SELECT customers.firstname, customers.lastname, SUM(total) AS total_purchases
FROM customers
LEFT JOIN invoices ON invoices.customerid = customers.customerid
GROUP BY customers.customerid
ORDER BY total_purchases DESC
LIMIT 1;
```
